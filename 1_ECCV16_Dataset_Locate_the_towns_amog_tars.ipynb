```python
def coors(filename):
  with open(filename) as fp:
    f = fp.read()
    f = f.split(' ')
    f = f[5:7]
    f = [float(x) for x in f]
  return tuple(f)
```

```python
with open('files.txt', 'r') as fp:
  file = fp.read()

all_files = file.split('\n')[1:-2]
```

```python
coordinates = dict()
for file in all_files:
  path = file

  subprocess.call(f'gsutil cp {file} . ', shell=True, executable='/bin/bash')

  filename = file.split('/')[-1]

  tar = tarfile.open(filename)

  members = tar.getmembers()

  members = [x for x in members if x.name[-3:] == 'txt']
  one = members[0]
  clear_output(wait=True)
  tar.extractall(members=[x for x in tar.getmembers() if x.name in [one.name]])

  for folder in os.listdir('.'):
    if folder[0] == '0' and '.' not in folder:
      for f in os.listdir(folder):
        coordinates[folder] = coors(f'{folder}/{f}')
  print(coordinates)

  subprocess.call(f'rm -rf {filename}', shell=True, executable='/bin/bash')
```
